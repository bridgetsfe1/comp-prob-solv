```python
##Step A: Importing packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
##Step A: Import Data
df = pd.read_csv("volume_pressure_data.csv")
df.head()
```

```python
##Step A: Data Exploration
mean_vp = df.mean()
median_vp = df.median()
sd_vp = df.std()

print("Mean:", mean_vp)
print("Median:", median_vp)
print("Standard Deviation:", sd_vp)
```

```python
##Step B: Data Visualization
df.plot("Volume", "Pressure", legend=False)


plt.xlabel("Volume")
plt.ylabel("Pressure")
plt.title("Volume vs. Pressure")
plt.grid()

plt.show()
```

```python
##Step C: Labeling Columns
vol = df['Volume'].values
pre = df['Pressure'].values
```

```python
##Step C: Curve Fitting
coefficients = np.polyfit(vol, pre, 2)
print(coefficients)
```

```python
##Step C: Curve Fitting to Previous Graph
best_fit = np.polyval(coefficients, vol)

df.plot("Volume", "Pressure", label="Original Data")
plt.plot(vol, best_fit, label="$y = .081x^2 + -6.93x + 180.08$")


plt.xlabel("Volume")
plt.ylabel("Pressure")
plt.title("Volume vs. Pressure")
plt.grid()
plt.legend()


plt.show()
```
