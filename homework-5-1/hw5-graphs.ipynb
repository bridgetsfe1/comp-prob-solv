```python
import numpy as np
import matplotlib.pyplot as plt
import overlap_H2 as fn
```

```python
#Plotting Random Sampling
np.random.seed(42)
n_points_list = [10, 10**2, 10**3, 10**4, 10**5, 10**6, 10**7, 9**8] #Kernel crahses for 10**8
L = 20
R = 2

S_R_rand = fn.random_sampling(n_points_list, L, R)

plt.errorbar(n_points_list, S_R_rand, fmt ='o-', label ="S(R)")
plt.xscale('log')
plt.xlim(1, 10**8)
plt.xlabel('Number of Points Sampled')
plt.ylabel('Overlap Integral Value')
plt.title('Overlap Integral of Two H 2$p_z$ Orbitals using Random Sampling')
plt.grid()
plt.legend()
plt.show()
print(S_R_rand)
```

```python
#Plotting Importance Sampling
np.random.seed(42)
S_R_imp = fn.important_sampling(n_points_list, R)

plt.figure(figsize=(8, 6))
plt.plot(n_points_list, S_R_imp, 'o-', label ="S(R)")

plt.xlabel('Number of Points Sampled')
plt.xscale('log')
plt.xlim(1, 10**8)
plt.ylabel('Overlap Integral Value')
plt.title('Overlap Integral of Two H 2$p_z$ Orbitals using Importance Sampling')
plt.grid()
plt.legend()
plt.show()
print(S_R_imp)
```

```python
#Plotting overlap integral as function of separation distance
R_values = np.arange(0, 20.5, 0.5)
S_R_values = []
for R in R_values:
    integral=fn.important_sampling(n_points_list=[10**6],R=R)
    S_R_values.append(integral)
# Plot S(R) vs. R
plt.plot(R_values, S_R_values, marker='o')
plt.xlabel('Separation Distance (R)')
plt.ylabel('Overlap Integral S(R)')
plt.title('S(R) vs. R')
plt.grid()
plt.show()
```
